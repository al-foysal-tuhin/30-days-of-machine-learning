**Data Clean**

```python
sl_new = sl.drop(['Unnamed: 0','salary_currency','salary_in_usd','employee_residence','salary'],axis='columns')
sl_new
```

```python
y= sl.salary
y
```

**One hot Encoding**

```python
x = pd.get_dummies(sl_new,drop_first=True)
x.head()
```

**Label Encoding**

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
```

```python
sl.job_title = label_encoder.fit_transform(sl.job_title)
sl.head()
```

**Train Test Split**

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=20)
```

```python
x_train
```

```python
x_train.shape
```

```python
y_train
```

```python
y_train.shape
```

**Linear Reggression**

```python
lreg = linear_model.LinearRegression()
lreg.fit(x_train, y_train)
```

```python
lreg.predict(x_test)
```

**Accuracy**

```python
lreg.score(x_test, y_test)
```

**Error**

Mean Square Error

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test,lreg.predict(x_test))
mse
```

Root Mean Square Error

```python
from math import sqrt
rmse = sqrt(mse)
rmse
```
